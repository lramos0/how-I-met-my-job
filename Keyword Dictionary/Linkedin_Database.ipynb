```python
import re
import ftfy
import pycountry
import pandas as pd

# --- Build lookups ---
NAME_TO_A2 = {}
for c in pycountry.countries:
    NAME_TO_A2[c.name.lower()] = c.alpha_2
    if hasattr(c, "official_name"):
        NAME_TO_A2[c.official_name.lower()] = c.alpha_2

# Extra aliases
NAME_TO_A2.update({
    "united states": "US", "u.s.": "US", "usa": "US", "u.s.a": "US",
    "united kingdom": "GB", "uk": "GB", "u.k.": "GB",
    "türkiye": "TR", "turkiye": "TR", "turkey": "TR",
    "south korea": "KR", "republic of korea": "KR",
    "north korea": "KP",
    "russia": "RU",
    "czechia": "CZ", "czech republic": "CZ",
    "viet nam": "VN"
})

A2 = {c.alpha_2 for c in pycountry.countries}
A3_TO_A2 = {c.alpha_3: c.alpha_2 for c in pycountry.countries}

US_STATES = {
    "AL","AK","AZ","AR","CA","CO","CT","DE","FL","GA","HI","ID","IL","IN","IA",
    "KS","KY","LA","ME","MD","MA","MI","MN","MS","MO","MT","NE","NV","NH","NJ",
    "NM","NY","NC","ND","OH","OK","OR","PA","RI","SC","SD","TN","TX","UT","VT",
    "VA","WA","WV","WI","WY","DC","PR"
}

NAME_PATTERNS = sorted(NAME_TO_A2.keys(), key=len, reverse=True)
NAME_PATTERNS = [(re.compile(rf"(?<!\w){re.escape(n)}(?!\w)", re.IGNORECASE), NAME_TO_A2[n]) for n in NAME_TO_A2]

TRAILING_CODE_RE = re.compile(r"[\s,()\-\u2013\u2014]+([A-Za-z]{2,3})\s*$")

def derive_country_code(location: str, existing_code: str | None = None) -> str | None:
    """Standardize country code from job_location + existing country_code"""
    if pd.notna(existing_code) and str(existing_code).strip():
        code = str(existing_code).strip().upper()
        if code in A2:
            return code

    if not isinstance(location, str) or not location.strip():
        return None

    text = ftfy.fix_text(location).strip()
    lower = text.lower()

    # 1) Match full country names
    for pat, code in NAME_PATTERNS:
        if pat.search(lower):
            return code

    # 2) US states → US
    tokens = re.findall(r"[A-Za-z]+", text.upper())
    if any(tok in US_STATES for tok in tokens):
        return "US"

    # 3) Trailing codes (", TR", "(IND)")
    m = TRAILING_CODE_RE.search(text)
    if m:
        trailing = m.group(1).upper()
        if trailing in A2:
            return trailing
        if trailing in A3_TO_A2:
            return A3_TO_A2[trailing]

    # 4) Standalone codes (avoid substrings like 'IS' in 'Istanbul')
    for tok in tokens:
        if tok in A2:
            return tok
        if tok in A3_TO_A2:
            return A3_TO_A2[tok]

    return None
```

```python
df["country_code_std"] = df.apply(
    lambda r: derive_country_code(r.get("job_location", ""), r.get("country_code")),
    axis=1
)

# Inspect problematic rows
print(df.loc[[1,2,3,4], ["job_location","country_code","country_code_std"]])
```

canonicalize job_industries and job_function so you have clean, consistent categories for analysis and matching.

```python
#Setup helpers (split, normalize, map synonyms)
import re
import pandas as pd

# Flexible multi-value separators: comma, semicolon, slash
SEP_RE = re.compile(r"\s*[,;/]\s*")

def split_normalize(cell: str):
    """Split a multi-valued string to a list of lowercased, trimmed tokens."""
    if pd.isna(cell) or str(cell).strip() == "":
        return []
    items = [x.strip().lower() for x in SEP_RE.split(str(cell)) if x.strip() != ""]
    # Normalize basic punctuation/spacing to underscores
    items = [re.sub(r"\s+", " ", x) for x in items]      # collapse spaces
    items = [x.replace("&", " and " ) for x in items]    # unify '&' → 'and'
    items = [re.sub(r"\s+", "_", x.strip()) for x in items]  # spaces → underscores
    return items

def dedupe_preserve_order(items):
    seen, out = set(), []
    for x in items:
        if x not in seen:
            seen.add(x); out.append(x)
    return out
```

```python
#Define synonym maps (edit/extend to your data)
# INDUSTRY synonyms → canonical labels
INDUSTRY_SYNONYMS = {
    "information_technology_and_services": "information_technology",
    "information_technology": "information_technology",
    "it_services": "information_technology",
    "computer_software": "software",
    "software": "software",
    "internet": "software",
    "financial_services": "finance",
    "banking": "finance",
    "capital_markets": "finance",
    "insurance": "insurance",
    "health_care": "healthcare",
    "healthcare": "healthcare",
    "biotechnology": "biotech",
    "biotech_research": "biotech",
    "medical_devices": "medical_devices",
    "pharmaceuticals": "pharma",
    "management_consulting": "consulting",
    "professional_services": "consulting",
    "retail": "retail",
    "e_commerce": "ecommerce",
    "ecommerce": "ecommerce",
    "education_management": "education",
    "e_learning": "education",
    "higher_education": "education",
    "telecommunications": "telecom",
    "semiconductors": "semiconductors",
    "electronics": "electronics",
    "automotive": "automotive",
    "logistics_and_supply_chain": "logistics",
    "transportation_trucking_railroad": "transportation",
}

# FUNCTION synonyms → canonical labels
FUNCTION_SYNONYMS = {
    "engineering": "engineering",
    "software_engineering": "engineering",
    "data_engineering": "data_engineering",
    "machine_learning_engineering": "ml_engineering",
    "devops": "devops",
    "site_reliability_engineering": "sre",
    "it": "it",
    "information_technology": "it",

    "data_science": "data_science",
    "data_analysis": "data_analytics",
    "analytics": "data_analytics",
    "business_analysis": "business_analytics",
    "business_analytics": "business_analytics",
    "bi": "business_intelligence",
    "business_intelligence": "business_intelligence",

    "product_management": "product_management",
    "project_management": "project_management",
    "program_management": "program_management",

    "marketing": "marketing",
    "digital_marketing": "marketing",
    "sales": "sales",
    "customer_success": "customer_success",
    "operations": "operations",
    "finance": "finance",
    "human_resources": "hr",
    "people_operations": "hr",
    "design": "design",
    "ux": "design",
    "ui_ux": "design",
}
```

```python
#Canonicalize both columns → list + string columns
#This produces:job_industries_std / job_function_std: lists & job_industries_std_str / job_function_std_str: comma-separated strings
def canonicalize_list(cell, synonyms):
    items = split_normalize(cell)
    # map synonyms; if no mapping, keep original
    items = [synonyms.get(x, x) for x in items]
    # dedupe while preserving order
    items = dedupe_preserve_order(items)
    return items

# Ensure columns exist
for col in ["job_industries", "job_function"]:
    if col not in df.columns:
        df[col] = pd.Series(dtype="object")

df["job_industries_std"] = df["job_industries"].apply(lambda x: canonicalize_list(x, INDUSTRY_SYNONYMS))
df["job_function_std"]  = df["job_function"].apply(lambda x: canonicalize_list(x, FUNCTION_SYNONYMS))

# Also provide string versions for easy storage/filtering
df["job_industries_std_str"] = df["job_industries_std"].apply(lambda lst: ", ".join(lst) if lst else pd.NA)
df["job_function_std_str"]   = df["job_function_std"].apply(lambda lst: ", ".join(lst) if lst else pd.NA)

df[["job_industries","job_industries_std_str","job_function","job_function_std_str"]].head(10)
```

```python
#Create long mapping tables + one-hot encodings
# LONG: explode to one row per (job_posting_id, value)
inds_long = (
    df[["job_posting_id","job_industries_std"]]
    .explode("job_industries_std")
    .dropna()
    .rename(columns={"job_industries_std":"industry"})
)

func_long = (
    df[["job_posting_id","job_function_std"]]
    .explode("job_function_std")
    .dropna()
    .rename(columns={"job_function_std":"function"})
)

# WIDE: one-hot features
inds_dummies = pd.crosstab(inds_long["job_posting_id"], inds_long["industry"]).add_prefix("ind_")
func_dummies = pd.crosstab(func_long["job_posting_id"], func_long["function"]).add_prefix("func_")

# Merge back (left-join so we keep all jobs)
df = df.merge(inds_dummies, left_on="job_posting_id", right_index=True, how="left")
df = df.merge(func_dummies, left_on="job_posting_id", right_index=True, how="left")

# Fill NaNs from dummy merge with 0
dummy_cols = [c for c in df.columns if c.startswith("ind_") or c.startswith("func_")]
df[dummy_cols] = df[dummy_cols].fillna(0).astype(int)

print("Added one-hot columns:", len(dummy_cols))
```
